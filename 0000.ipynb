- kind가 running이며 pulse가 100 이상인  조건 필터링
- 시간이 가장 긴 순서로 내림차순 정렬
- 상위 10개 출력

```python
cond1 = exercise['kind'] == 'running'
cond2 = exercise['pulse'] >= 100
exercise.loc[cond1 & cond2].sort_values('time', ascending=False).head(10)
```

- 운동시간이 30분이상인 걷기운동
- no fat이고 펄스가 100이상인 달리기
- id 출력

```python



cond3 = exercise['diet'] == 'no fat'
cond4 = exercise['pulse'] >= 100
cond5 = exercise['kind'] == 'running'

exercise.loc[cond3 & cond4 & cond5, ['id']]
```
