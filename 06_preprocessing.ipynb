```python
df_copy['운동량'] = pd.to_numeric(df_copy['운동량'], errors='coerce')
```

```python
df_copy.info()
```

### cut()

```python
bins = [1, 100, 200, 300, df_copy['운동량'].max()]
pd.cut(df_copy['운동량'], bins)
```

```python
pd.cut(df_copy['운동량'], bins=10)
pd.cut(df_copy['운동량'], bins=10).value_counts() # 균등한경우
```

```python
pd.qcut(df_copy['운동량'], q=10)
pd.qcut(df_copy['운동량'], q=10).value_counts() # 비율
```

```python
qcut_bins = [0, 0.2, 0.8, 1]
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
df_copy.head(2)
# pd.qcut(df_copy['운동량'], qcut_bins, labels=['적음', '보통', '많음']).value_counts()
# pd.qcut(df_copy['운동량'], qcut_bins).value_counts()
```
