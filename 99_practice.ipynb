```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
df
```

```python
df.values
```

```python
df.sort_values('body_mass_g', ascending=False)
```

```python
# 5000g 이상의 펭귄 출력
```

```python
cond = df['body_mass_g'] >= 5000

df.loc[cond].sort_values('body_mass_g', ascending=False)
```

```python
# Gentoo종의 flipper_length 평균
```

```python
cond = df['species'] == 'Gentoo'
df.loc[cond, 'flipper_length_mm'].mean()
```

```python
# Female Chinstrap종의 bill_length 높은순 정렬
```

```python
cond1 = df['sex'] == 'Female'
cond2 = df['species'] == 'Chinstrap'

df.loc[cond1 & cond2].sort_values('bill_length_mm', ascending=False).head(10)
```

```python
# species(종)별 성별 구분하여 평균 몸무게 출력
# => ex) Adelie, Female / Adelie, male / Gentoo, Female...
```

```python
df['species'].value_counts()
```

```python
cond1 = df['species'] == 'Adelie'
cond2 = df['sex'] == 'Male'

Adelie_male_mean = df.loc[cond1 & cond2, 'body_mass_g'].mean()
print(Adelie_male_mean)
```

```python
cond3 = df['species'] == 'Adelie'
cond4 = df['sex'] == 'Female'

Adelie_female_mean = df.loc[cond3 & cond4, 'body_mass_g'].mean()
print(Adelie_female_mean)
```

```python
cond5 = df['species'] == 'Gentoo'
cond6 = df['sex'] == 'Male'

Gentoo_male_mean = df.loc[cond5 & cond6, 'body_mass_g'].mean()
print(Gentoo_male_mean)
```
